```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("anJWfQsT0ZKuEBip") as m:
    # Structure for material id
    struct = m.get_structure_by_material_id("mp-961661")



xrd = XRDCalculator(wavelength='CuKa')
pattern = xrd.get_pattern(struct)
```

```python
print("Premier pic :\n")
print("Angle 2\u0398 = ", pattern.x[0],"°")
print("Intensité = ", pattern.y[0],"%")
print("Indices hkl = ", pattern.hkls[0], '\n')
print("--------------- \n")

print("Deuxième pic :\n")
print("Angle 2\u0398 = ", pattern.x[1],"°")
print("Intensité = ", pattern.y[1],"%")
print("Indices hkl = ", pattern.hkls[1],'\n')
print("--------------- \n")

print("Troisième pic :\n")
print("Angle 2\u0398 = ", pattern.x[2],"°")
print("Intensité = ", pattern.y[2],"%")
print("Indices hkl = ", pattern.hkls[2],'\n')

plot = xrd.get_plot(struct, two_theta_range=(0,45))
plot.show()
```
